```python
# Merging datasets

df_merge_countries = pd.merge(df_new, df_regions, how='left', on='country')

df_merge_countries.head()
```

```python
df_merged = df_merge_countries.copy()

display(df_merged.head(5))
display(df_merged.tail(5))
```

```python
# Checking null values

df_merged.isnull().sum()
```

```python
# Checking NaN values

df_merged.isna().sum()
```

```python
# Grouping by region

region_pop_gby = df_merged.groupby(['region','year']).aggregate(np.sum).reset_index()
```

```python
region_pop_gby.head(3)
```

```python
# Renaming regions

region_pop_gby = region_pop_gby.replace({
    'ASIA (EX. NEAR EAST)' : 'East & Southeast Asia',
    'BALTICS' : 'Baltics',
    'C.W. OF IND. STATES' : 'CIS',
    'EASTERN EUROPE' : 'Eastern Europe',
    'LATIN AMER. & CARIB' : 'Latin America & Caribbean',
    'NEAR EAST' : 'Middle East',
    'NORTHERN AFRICA' : 'North Africa',
    'NORTHERN AMERICA' : 'North America',
    'OCEANIA' : 'Oceania',
    'SUB-SAHARAN AFRICA' : 'Sub-Saharan Africa',
    'WESTERN EUROPE' : 'Western Europe'
    }, regex=False)
```

```python
# Checking regions unique values

region_pop_gby['region'].unique()
```

## **3. Criação do Gráfico**

```python
# Creating graph

fig_scatter = px.scatter(region_pop_gby,x='year',
                         y='population', color= 'region',
                         animation_frame='year', size='population',
                         animation_group='region', range_x=[1980,2099],
                         range_y=[min(region_pop_gby['population']),
                                  max(region_pop_gby['population'])])
```

```python
fig_bar = px.bar(region_pop_gby, x = 'region',
       y = 'population', color= 'region',
       animation_frame = 'year',
       range_y = [min(region_pop_gby['population']),
                  max(region_pop_gby['population'])],
       title = 'Regions total population per year')
```
